```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException, NoSuchElementException
import pandas as pd
import time

driver_KE = webdriver.Chrome()
driver_KE.get("https://www.redbus.in/online-booking/ksrtc-kerala/?utm_source=rtchometile")
time.sleep(3)
driver_KE.maximize_window()
wait = WebDriverWait(driver_KE, 20)

def KE_link_route(path):
    LINKS_KE = []
    ROUTE_KE = []

    try:
        # Get total number of pages dynamically
        total_pages = int(driver_KE.find_element(By.XPATH, '//*[@class="DC_117_paginationTable"]//div[last()]').text)
        for i in range(1, total_pages + 1):
            if len(driver_KE.window_handles) == 0:
                raise NoSuchWindowException("Browser window was closed or lost.")

            # Retrieve route links and names
            paths = driver_KE.find_elements(By.XPATH, path)
            for link in paths:
                LINKS_KE.append(link.get_attribute("href"))
                ROUTE_KE.append(link.text)

            # Handle pagination
            if i < total_pages:
                try:
                    # Wait for the "Next" button to become clickable
                    next_button = wait.until(EC.element_to_be_clickable((
                        By.XPATH, f'//div[@class="DC_117_pageTabs " and text()={i+1}]'
                    )))
                    
                    # Scroll to the button to ensure visibility
                    driver_KE.execute_script("arguments[0].scrollIntoView(true);", next_button)
                    time.sleep(1)

                    # Attempt to click the button
                    next_button.click()
                except Exception as e:
                    print(f"Error clicking on page {i + 1}: {e}")
                    break

    except Exception as e:
        print(f"Error occurred: {e}")
        return [], []

    return LINKS_KE, ROUTE_KE

LINK_KE, ROUTE_KE = KE_link_route("//a[@class='route']")

# Create a DataFrame
df_k = pd.DataFrame({"Route_name": ROUTE_KE, "ROUTE_link": LINK_KE})
# Close the browser
driver_KE.quit()
```
